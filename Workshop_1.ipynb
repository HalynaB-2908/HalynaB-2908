```python
stud_DataFrame = pd.read_excel('2_workshop_DataSet.xlsx', header=0)
```

```python
total_count = stud_DataFrame[
    (stud_DataFrame["Subject1"].isin(["excellent", "good"])) &
    (stud_DataFrame["Subject2"].isin(["excellent", "good"]))
].shape[0]

result_total = pd.DataFrame({
    "Group": ["All"],
    "Number of Students with excellent and good Marks": [total_count]
})

display(result_total)
```

```python
Stud_Unsat = stud_DataFrame[
    (stud_DataFrame["Subject1"].isin(["Unsatisfactory (with the exam retake option)","Unsatisfactory (with mandatory repetition of the course)"])) &
    (stud_DataFrame["Subject2"].isin(["Unsatisfactory (with the exam retake option)","Unsatisfactory (with mandatory repetition of the course)"]))
].groupby("Group").size().reset_index(name="Number of Students with Unsatisfactory Marks")

display(Stud_Unsat)
```

```python
grade_labels = [
    "excellent",
    "good",
    "satisfactory",
    "Unsatisfactory (with the exam retake option)",
    "Unsatisfactory (with mandatory repetition of the course)"
]

grade_counts = {}

for grade in grade_labels:
    grade_counts[grade] = stud_DataFrame[
        (stud_DataFrame["Subject1"] == grade) & (stud_DataFrame["Subject2"] == grade)
    ].groupby("Group").size()

grade_summary = pd.DataFrame(grade_counts).fillna(0).reset_index()

grade_summary.loc[len(grade_summary)] = ["Total"] + grade_summary[grade_labels].sum().tolist()

display(grade_summary)
```

```python
unsatisfactory_labels = [
    "Unsatisfactory (with the exam retake option)",
    "Unsatisfactory (with mandatory repetition of the course)"
]

groups_with_unsat = stud_DataFrame[
    stud_DataFrame["Subject1"].isin(unsatisfactory_labels) |
    stud_DataFrame["Subject2"].isin(unsatisfactory_labels)
]["Group"].unique()

groups_without_unsat = stud_DataFrame.loc[~stud_DataFrame["Group"].isin(groups_with_unsat), "Group"].unique()

groups_without_unsat_df = pd.DataFrame(groups_without_unsat, columns=["Groups without Unsatisfactory Marks"])

if groups_without_unsat_df.empty:
    groups_without_unsat_df.loc[0] = 0

display(groups_without_unsat_df)
```

```python
np.random.seed(42)
exam1_start = pd.to_datetime("2023-12-04") + pd.to_timedelta(np.random.randint(8, 11, size=len(stud_DataFrame)), unit="h")
exam1_finish = exam1_start + pd.to_timedelta(np.random.randint(30, 120, size=len(stud_DataFrame)), unit="m")

exam2_start = exam1_start + pd.to_timedelta(np.random.randint(1, 3, size=len(stud_DataFrame)), unit="h")
exam2_finish = exam2_start + pd.to_timedelta(np.random.randint(30, 120, size=len(stud_DataFrame)), unit="m")

# Додаємо нові стовпці до DataFrame
stud_DataFrame["Exam1 Start Time"] = exam1_start
stud_DataFrame["Exam1 Finish Time"] = exam1_finish
stud_DataFrame["Exam2 Start Time"] = exam2_start
stud_DataFrame["Exam2 Finish Time"] = exam2_finish

# Обчислюємо тривалість кожного екзамену (у хвилинах)
stud_DataFrame["Exam1 Duration"] = (stud_DataFrame["Exam1 Finish Time"] - stud_DataFrame["Exam1 Start Time"]).dt.total_seconds() / 60
stud_DataFrame["Exam2 Duration"] = (stud_DataFrame["Exam2 Finish Time"] - stud_DataFrame["Exam2 Start Time"]).dt.total_seconds() / 60

# Обчислюємо середній час тестування для кожного студента
stud_DataFrame["Average Test Duration"] = (stud_DataFrame["Exam1 Duration"] + stud_DataFrame["Exam2 Duration"]) / 2

# Сортуємо студентів за середнім часом тестування
sorted_df = stud_DataFrame.sort_values(by="Average Test Duration")

# Отримуємо 5 студентів з найкоротшим середнім часом тестування
shortest_tests = sorted_df.head(5)

# Отримуємо 5 студентів з найдовшим середнім часом тестування
longest_tests = sorted_df.tail(5)

# Відображаємо результати
display(stud_DataFrame)
display(shortest_tests)
display(longest_tests)
```
