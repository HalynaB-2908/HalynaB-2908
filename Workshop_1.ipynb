```python
print ("Adding a new column using the existing colums in DataFrame:")
df2['Total_score_Sub1']=df2['Subject1. Test 1']+df2['Subject1. Test2']
df2.head(10)
```

```python
print ("Adding a new column using the existing colums in DataFrame:")
df2['Total_score_Sub2']=df2['Subject2. Test 1']+df2['Subject2. Test 2']
df2.head(10)
```

```python
df2.loc[df2['Total_score_Sub1']>=90, 'ECTS_grading_scale_Sub1']='A'
df2.loc[(df2['Total_score_Sub1']>=82) & (df2['Total_score_Sub1']<=89), 'ECTS_grading_scale_Sub1']='B'
df2.loc[(df2['Total_score_Sub1']>=74) & (df2['Total_score_Sub1']<=81), 'ECTS_grading_scale_Sub1']='C'
df2.loc[(df2['Total_score_Sub1']>=64) & (df2['Total_score_Sub1']<=73), 'ECTS_grading_scale_Sub1']='D'
df2.loc[(df2['Total_score_Sub1']>=60) & (df2['Total_score_Sub1']<=63), 'ECTS_grading_scale_Sub1']='E'
df2.loc[(df2['Total_score_Sub1']>=35) & (df2['Total_score_Sub1']<=59), 'ECTS_grading_scale_Sub1']='FX'
df2.loc[df2['Total_score_Sub1']<=34, 'ECTS_grading_scale_Sub1']='F'
df2.head(10)
```

```python
df2.loc[df2['Total_score_Sub2']>=90, 'ECTS_grading_scale_Sub2']='A'
df2.loc[(df2['Total_score_Sub2']>=82) & (df2['Total_score_Sub2']<=89), 'ECTS_grading_scale_Sub2']='B'
df2.loc[(df2['Total_score_Sub2']>=74) & (df2['Total_score_Sub2']<=81), 'ECTS_grading_scale_Sub2']='C'
df2.loc[(df2['Total_score_Sub2']>=64) & (df2['Total_score_Sub2']<=73), 'ECTS_grading_scale_Sub2']='D'
df2.loc[(df2['Total_score_Sub2']>=60) & (df2['Total_score_Sub2']<=63), 'ECTS_grading_scale_Sub2']='E'
df2.loc[(df2['Total_score_Sub2']>=35) & (df2['Total_score_Sub2']<=59), 'ECTS_grading_scale_Sub2']='FX'
df2.loc[df2['Total_score_Sub2']<=34, 'ECTS_grading_scale_Sub2']='F'
df2.head(10)
```

```python
df2.loc[df2['ECTS_grading_scale_Sub1']=='A', 'National_grading_scale_Sub1']='excellent'
df2.loc[(df2['ECTS_grading_scale_Sub1']=='B') | (df2['ECTS_grading_scale_Sub1']=='C'), 'National_grading_scale_Sub1']='good'
df2.loc[(df2['ECTS_grading_scale_Sub1']=='D') | (df2['ECTS_grading_scale_Sub1']=='E'), 'National_grading_scale_Sub1']='satisfactory'
df2.loc[df2['ECTS_grading_scale_Sub1']=='FX', 'National_grading_scale_Sub1']='Unsatisfactory (with the exam retake option)'
df2.loc[df2['ECTS_grading_scale_Sub1']=='F', 'National_grading_scale_Sub1']='Unsatisfactory (with mandatory repetition of the course)'
df2.head(10)
```

```python
df2.loc[df2['ECTS_grading_scale_Sub2']=='A', 'National_grading_scale_Sub2']='excellent'
df2.loc[(df2['ECTS_grading_scale_Sub2']=='B') | (df2['ECTS_grading_scale_Sub2']=='C'), 'National_grading_scale_Sub2']='good'
df2.loc[(df2['ECTS_grading_scale_Sub2']=='D') | (df2['ECTS_grading_scale_Sub2']=='E'), 'National_grading_scale_Sub2']='satisfactory'
df2.loc[df2['ECTS_grading_scale_Sub2']=='FX', 'National_grading_scale_Sub2']='Unsatisfactory (with the exam retake option)'
df2.loc[df2['ECTS_grading_scale_Sub2']=='F', 'National_grading_scale_Sub2']='Unsatisfactory (with mandatory repetition of the course)'
df2.head(10)
```
